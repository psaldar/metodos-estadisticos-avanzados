# Consolidacion de la Informacion

```python
import pandas as pd
```

```python
info_ind = pd.read_csv(r"C:\Users\pasal\Universidad EAFIT\Ing Mats in Data Science - General\datasets\indicadores_anual.csv", sep = ';')
info_gv = pd.read_csv(r"C:\Users\pasal\Google Drive\EAFIT\Ciencia de Datos y Analitica\02 Segundo Semestre\Metodos Estadisticos Avanzados\Trabajo Final\metodos-estadisticos-avanzados\PreparacionDatos\Costos_Gastos_ventas.csv")
info_gv = info_gv.rename(columns = {'Año':'YEAR'})
info_gv['YEAR'] = info_gv['YEAR'].astype(int)
```

```python
info_ind['YEAR'] = info_ind['PERIODO'].apply(lambda x: x.split('/')[2])
info_ind['YEAR'] = info_ind['YEAR'].astype(int)
info_indicadores = pd.pivot_table(info_ind, index = 'YEAR', values = 'VALOR', columns = 'INDICADOR', aggfunc='sum').reset_index()
```

```python
info_indicadores
```

```python
all_data = info_gv.merge(info_indicadores, how = 'left', on = 'YEAR').drop(columns = ['tasa ocupacion'])
```

```python
new_cols = []
for col in all_data.columns.values:
    new_cols.append(col.replace(' ','_'))
    
all_data.columns = new_cols
```

```python
all_data.to_csv('Data/consolidado_info.csv', index = False, sep = ',')
all_data.to_csv('C:/Users/pasal/Universidad EAFIT/Ing Mats in Data Science - General/datasets/consolidado_info.csv', index = False, sep = ',')
```
